```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
```

```python
DATADIR = "C:\Tutorial\Machine-Learning\Datasets\Cat vs Dog"
CATEGORIES = ["Cat","Dog"]

for category in CATEGORIES :
    path = os.path.join(DATADIR,category)    # path to Cat or Dog dir
    for img in os.listdir(path):
        img_path = os.path.join(path,img)
        img_array = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
```

```python
IMG_SIZE = 100

new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
training_data = []
IMG_SIZE = 100

for category in CATEGORIES :
    path = os.path.join(DATADIR,category)    # path to Cat or Dog dir
    class_num = CATEGORIES.index(category)   # for 0 or 1 association with image
    for img in os.listdir(path):
        try:                                 # try block coz some of the images are broken
            img_path = os.path.join(path,img)
            img_array = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
            new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
            training_data.append([new_array,class_num]) 
        except Exception as e:
            pass
```
