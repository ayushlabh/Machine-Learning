```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# If a column of table contains index , use 'index_col'

review = pd.read_csv("C:\Tutorial\Machine-Learning\Datasets\Wine Review\winemag-data_first150k.csv",index_col = 0)
```

```python
review.head()
```

```python
review.shape
```

## 1. Subplotting

```python
plt.subplots(2, 1,figsize = (14,10))
```

## 1. Bar Graph

```python
review['province'].value_counts().head(10).plot.bar()
```

```python
(review['province'].value_counts().head(10) / len(review)).plot.bar()
```

```python
review['points'].value_counts().sort_index().plot.bar()
```

## 2. Line Graph

```python
review['points'].value_counts().sort_index().plot.line()
```

## 3. Area Graph

```python
review['points'].value_counts().sort_index().plot.area()
```

## 4. Histogram

```python
review[review['price'] < 200]['price'].plot.hist()
```

## 5. Scatter Plot

```python
review.plot.scatter(x='price', y='points')
```

```python
review[review['price']<100].sample(100).plot.scatter(x='price',y='points')
```

## 6. Hexplot

```python
review[review['price'] < 100].plot.hexbin(x='price', y='points')
```

```python
review[review['price'] < 100].plot.hexbin(x='price', y='points',gridsize = 15)
```
