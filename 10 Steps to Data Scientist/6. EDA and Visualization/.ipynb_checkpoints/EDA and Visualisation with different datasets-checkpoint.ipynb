# 1. Student Performance dataset

Feature Classification 

1. Categorical : gender , race/ethnicity , parental level of education , lunch , test preparation course
<br><br>
2. Numerical : math score , reading score , writing score

```python
# Categories in categorical data

print("Gender:",train["gender"].unique())
print()

print("Race/Ethnicity:",train["race/ethnicity"].unique())
print()

print("Parental level of education:",train["parental level of education"].unique())
print()

print("Test preparation course:",train["test preparation course"].unique())

```

```python
# Value counts in categorical data

print(train["gender"].value_counts())
print()

print(train["race/ethnicity"].value_counts())
print()

print(train["parental level of education"].value_counts())
print()

print(train["test preparation course"].value_counts())
```

```python
# Bargraph for categorical data

sns.set(style="whitegrid")

gender = train["gender"].value_counts()

sns.barplot(x = gender.index, y = gender.values,hue=['female','male'])
plt.legend(loc=9)
plt.show()


labels = train["race/ethnicity"].value_counts().index
values = lebels = train["race/ethnicity"].value_counts().values

plt.figure(figsize=(7,7))
plt.pie(values , labels = labels,autopct = "%1.1f%%")
plt.title("Race/Ethnicity distribution")
plt.show()


labels = train["parental level of education"].value_counts().index
values = train["parental level of education"].value_counts().values


sns.barplot(x=labels, y=values , estimator=lambda x: sum(x)*100/len(train))
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.show()

sns.countplot(train["test preparation course"],hue = train["gender"])
plt.show()
```

```python
# Bargraph with division in category

sns.barplot(x="parental level of education" , y = "writing score" , data=train)
plt.xticks(rotation=45)
plt.show()

sns.barplot(x="parental level of education" , y = "math score" , hue="gender" , data=train)
plt.legend()
plt.xticks(rotation=30)
plt.show()

sns.barplot(x="gender" , y = "reading score" , hue = "race/ethnicity" , data=train)
plt.legend()
plt.show()
```

```python
# Correlation with heatmap

corr = train.corr()

sns.heatmap(corr,vmin=0,vmax=1,annot = True)
plt.show()
```

```python
# Pair plot for Numerical data

sns.pairplot(train)
plt.show()
```
